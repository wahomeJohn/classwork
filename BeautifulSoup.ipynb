```python
import requests
url = 'https://www.brainyquote.com/topics/motivational-quotes'
r = requests.get(url)
print(r.content)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(r.content,'html5lib')
quotes = []
```

```python
table = soup.find('div', attrs ={'id':'quotesList'})
print(table.prettify())
```

```python
for row in table.findAll('div', attrs = {'class':'qti-listm'}):
    print(row)
```

```python
for row in table.findAll('div', attrs = {'class':'qti-listm'}):
  quote ={} # we fill this in a loop
  try:
      quote['author'] = row.img['alt'].split("-")[1]
      quote['text'] = row.img['alt'].split("-")[0]
      quotes.append(quote)

      #print (quote)
  except TypeError:
    continue
 
```

```python
import csv
```

```python
filename = 'short_quotes.csv'
with open(filename, 'w', newline='') as f:
    w = csv.DictWriter(f,['author','text'])
    w.writeheader()
    for quote in quotes:
        w.writerow(quote)
```
